```python
import os
import subprocess
import whisper
import pandas as pd
import shutil
import random  # Para gerar n√∫meros aleat√≥rios

# Caminho para a planilha
caminho_planilha = "/home/pc/Downloads/Dataset elei√ß√µes municipais 2024 - Candidatos prefeituras capitais (Instagram) - Dataset.csv"

# Caminho para a planilha de erros
caminho_planilha_erros = "/home/pc/Downloads/Erros_Processamento_Instagram_11_sabemos.xlsx"
nome_aba_erros = "Erros"

# Fun√ß√£o para ler os dados da planilha
def ler_planilha():
    df = pd.read_csv(caminho_planilha)
    return df

# Fun√ß√£o para salvar os erros na planilha de erros
def salvar_erro(url, erro, primeira_vez=False):
    dados_erro = {"Link": [url], "Erro": [str(erro)]}
    df_erro = pd.DataFrame(dados_erro)
    
    if os.path.exists(caminho_planilha_erros) and not primeira_vez:
        df_existente = pd.read_excel(caminho_planilha_erros, sheet_name=nome_aba_erros)
        df_final = pd.concat([df_existente, df_erro], ignore_index=True)
    else:
        df_final = df_erro
    
    with pd.ExcelWriter(caminho_planilha_erros, engine="openpyxl", mode="w") as writer:
        df_final.to_excel(writer, sheet_name=nome_aba_erros, index=False)

# Fun√ß√£o para baixar o √°udio
def baixar_audio(url, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    audio_path = os.path.join(output_dir, "audio.mp3")  # Ou .wav se preferir
    comando = ["yt-dlp", "--extract-audio", "--audio-format", "mp3", "-o", audio_path, "--cookies-from-browser", "chrome", url]
    subprocess.run(comando, check=True)

    print(f"√Åudio baixado com sucesso: {audio_path}")
    return audio_path

# Fun√ß√£o para transcrever o √°udio usando Whisper
def transcrever_audio(audio_path):
    model = whisper.load_model("medium")
    result = model.transcribe(audio_path)
    
    return result['text']

# Fun√ß√£o para garantir que o nome do arquivo seja √∫nico
def gerar_nome_arquivo(nome_do_arquivo, pasta_destino):
    while os.path.exists(os.path.join(pasta_destino, f"{nome_do_arquivo}.txt")):
        nome_do_arquivo = f"{nome_do_arquivo}_{random.randint(1000, 9999)}"
    return nome_do_arquivo

# Fun√ß√£o principal
def main():
    df = ler_planilha()
    
    pasta_instagram = "/home/pc/Downloads/Instagram"
    if not os.path.exists(pasta_instagram):
        os.makedirs(pasta_instagram)

    primeira_vez = True
    
    for index, row in df.iterrows():
        url = row['Link']
        # Verifica se o corpo de texto est√° vazio e, caso esteja, gera um n√∫mero aleat√≥rio
        if pd.notnull(row['Corpo de texto']):
            nome_do_arquivo = str(row['Corpo de texto'])
        else:
            nome_do_arquivo = str(random.randint(1000, 9999))
        
        nome_do_arquivo = nome_do_arquivo[:50] if len(nome_do_arquivo) > 50 else nome_do_arquivo
        user_id_nome_pasta = str(row['ID Usu√°rio'])

        try:
            output_dir = os.path.join(pasta_instagram, user_id_nome_pasta)
            if not os.path.exists(output_dir):
                os.makedirs(output_dir)

            # Baixar o √°udio
            audio_path = baixar_audio(url, output_dir)
            print(f"Transcrevendo o √°udio {audio_path}...")
            transcricao = transcrever_audio(audio_path)

            # Garante que o nome do arquivo seja √∫nico
            nome_arquivo = gerar_nome_arquivo(nome_do_arquivo, output_dir)
            
            with open(os.path.join(output_dir, f"{nome_arquivo}.txt"), "w", encoding="utf-8") as f:
                f.write(transcricao)

            print(f"Transcri√ß√£o salva em: {os.path.join(output_dir, f'{nome_arquivo}.txt')}")
            os.remove(audio_path)

        except Exception as e:
            print(f"Erro ao processar o link {url}: {e}")
            salvar_erro(url, e, primeira_vez)
            primeira_vez = False

# Executar a fun√ß√£o principal
if __name__ == "__main__":
    main()
```
